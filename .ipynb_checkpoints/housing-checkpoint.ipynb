```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
pd.set_option('display.max_columns', 500)
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
test.head()
```

```python
train.head()
```

```python
fig, ax = plt.subplots(1, figsize = [12.5, 7])

ax.hist(train['SalePrice'], bins=100)

plt.show()
```

```python
fig, ax = plt.subplots(1, figsize = [12.5, 7])

ax.hist(train['LotArea'], bins=100)

plt.show()
```

```python
fig, ax = plt.subplots(1, figsize = [12.5, 7])

ax.scatter(train['LotArea'], train['SalePrice'])

plt.show()
```

```python
subclass_group = train.groupby('MSSubClass').agg({'MSSubClass': 'count'}).rename(columns={'MSSubClass': 'Count'}).reset_index()
subclass_group['MSSubClass'] = subclass_group['MSSubClass'].astype('str')

fig, ax = plt.subplots(1, figsize = [12.5, 7])

ax.bar(subclass_group['MSSubClass'], subclass_group['Count'])

plt.show()
```

```python
sns.boxplot(train['MSSubClass'], train['SalePrice'])
```
